```python
# Grabbed from Brian M. to use time midpoints, not end periods
def cesm_correct_time(ds):
    """Given a Dataset, check for time_bnds,
       and use avg(time_bnds) to replace the time coordinate.
       Purpose is to center the timestamp on the averaging inverval.   
       NOTE: ds should have been loaded using `decode_times=False`
    """
    assert 'time_bnds' in ds
    assert 'time' in ds
    correct_time_values = ds['time_bnds'].mean(dim='nbnd')
    # copy any metadata:
    correct_time_values.attrs = ds['time'].attrs
    ds = ds.assign_coords({"time": correct_time_values})
    ds = xr.decode_cf(ds)  # decode to datetime objects
    return ds

# - - - - - - - - - - - - - - - 
# Regrid utility 
# - - - - - - - - - - - - - - - 
def regrid_data(fromthis, tothis, method=1):
    """Regrid data using various different methods"""

    #Import necessary modules:
    import xarray as xr

    if method == 1:
        # kludgy: spatial regridding only, seems like can't automatically deal with time
        if 'time' in fromthis.coords:
            result = [fromthis.isel(time=t).interp_like(tothis) for t,time in enumerate(fromthis['time'])]
            result = xr.concat(result, 'time')
            return result
        else:
            return fromthis.interp_like(tothis)
    elif method == 2:
        newlat = tothis['lat']
        newlon = tothis['lon']
        coords = dict(fromthis.coords)
        coords['lat'] = newlat
        coords['lon'] = newlon
        return fromthis.interp(coords)
    elif method == 3:
        newlat = tothis['lat']
        newlon = tothis['lon']
        ds_out = xr.Dataset({'lat': newlat, 'lon': newlon})
        regridder = xe.Regridder(fromthis, ds_out, 'bilinear')
        return regridder(fromthis)
    elif method==4:
        # geocat
        newlat = tothis['lat']
        newlon = tothis['lon']
        result = geocat.comp.linint2(fromthis, newlon, newlat, False)
        result.name = fromthis.name
        
        return result
```

```python
class NetCDFAnalyzer:
    def __init__(self, dataDir, caseNames, preprocess_lnd=None, preprocess_atm=None,
                loading_method='optimized', time_chunks=12, spatial_chunks=None):
        """
        Initialize the analyzer with land and atmosphere netCDF files
        
        Parameters:
        lnd_filepath (str): Path to your land netCDF file
        atm_filepath (str): Path to your atmosphere netCDF file
        """
        self.dataDir = dataDir
        self.caseNames = caseNames

        # Generate file lists
        self.listFiles_lnd = np.sort(glob.glob(dataDir + caseNames + '/lnd/hist/*clm2.h1.????-*'))
        self.listFiles_atm = np.sort(glob.glob(dataDir + caseNames + '/atm/hist/*cam.h1.????-*'))
        
        print(f"Found {len(self.listFiles_lnd)} land files")
        print(f"Found {len(self.listFiles_atm)} atmosphere files")

        # Choose loading method
        if loading_method == 'lazy':
            load_func = self._load_dataset_lazy
        elif loading_method == 'optimized':
            load_func = lambda files, preproc: self._load_dataset_optimized(
                files, preproc, time_chunks, spatial_chunks)
        else:
            load_func = self._load_dataset
        
        # Load datasets using open_mfdataset
        print("Loading land dataset...")
        start_time = pd.Timestamp.now()
        self.lnd_ds = load_func(self.listFiles_lnd, preprocess_lnd)
        lnd_time = pd.Timestamp.now() - start_time
        print(f"Land dataset loaded in {lnd_time.total_seconds():.1f} seconds")
        
        print("Loading atmosphere dataset...")
        start_time = pd.Timestamp.now()
        self.atm_ds = load_func(self.listFiles_atm, preprocess_atm)
        atm_time = pd.Timestamp.now() - start_time
        print(f"Atmosphere dataset loaded in {atm_time.total_seconds():.1f} seconds")

        # Annoyingly, lnd and atm grids aren't always the same, 
        # Lat can differ on order of 1e-7, so just reassign... 
        self.lnd_ds['lat'] = self.atm_ds['lat']
        
        # Store dataset references
        self.datasets = {'lnd': self.lnd_ds, 'atm': self.atm_ds}
        
        # Get variables from both files
        self.lnd_variables = self._get_variables(self.lnd_ds)
        self.atm_variables = self._get_variables(self.atm_ds)
        
        # Create combined variable dictionary
        self.variables = {
            'lnd': self.lnd_variables,
            'atm': self.atm_variables
        }
        
        self.seasons = ['DJF', 'MAM', 'JJA', 'SON']  # Standard meteorological seasons

        print("Initialization complete!")
        print(f"Land variables: {len(self.lnd_variables)}")
        print(f"Atmosphere variables: {len(self.atm_variables)}")
        if loading_method == 'lazy':
            print("Note: Data is lazily loaded. Computation will happen when plots are generated.")
    
    
    def _load_dataset(self, file_list, preprocess_func=None):
        """
        Load multiple netCDF files into a single dataset
        
        Parameters:
        file_list (list): List of file paths
        preprocess_func (function, optional): Preprocessing function
        
        Returns:
        xarray.Dataset: Combined dataset
        """
        if len(file_list) == 0:
            raise ValueError("No files found matching the pattern")
        
        print(f"Loading {len(file_list)} files...")
        
        # Optimization 1: Use chunks and engine specifications
        dataset = xr.open_mfdataset(
            file_list,
            preprocess=preprocess_func,
            concat_dim='time',
            combine='nested',
            decode_times=False,
            data_vars='minimal',
            coords='minimal',  # Only load minimal coordinates
            parallel=True,
            engine='netcdf4',  # Specify engine explicitly
            # chunks={'time': 12},  # Chunk by time (adjust based on your data)
            # Alternative chunking strategies:
            chunks={'time': 'auto'}  # Let xarray decide
            # chunks={'time': 6, 'lat': 50, 'lon': 50}  # Multi-dimensional chunking
        )
        
        return dataset
        
    def _load_dataset_lazy(self, file_list, preprocess_func=None):
        """
        Alternative lazy loading approach - only loads metadata initially
        Call .load() or .compute() when you actually need the data
        """
        if len(file_list) == 0:
            raise ValueError("No files found matching the pattern")
        
        print(f"Lazy loading {len(file_list)} files (metadata only)...")
        
        dataset = xr.open_mfdataset(
            file_list,
            preprocess=preprocess_func,
            concat_dim='time',
            combine='nested',
            decode_times=False,
            data_vars='minimal',
            coords='minimal',
            parallel=True,
            engine='netcdf4',
            compat='override',
            chunks={'time': 'auto'},  # Let dask handle chunking
        )
        
        return dataset
    
    def _load_dataset_optimized(self, file_list, preprocess_func=None, 
                               time_chunk_size=12, spatial_chunks=None):
        """
        Highly optimized loading with custom chunk sizes and caching
        
        Parameters:
        file_list (list): List of file paths
        preprocess_func (function, optional): Preprocessing function
        time_chunk_size (int): Number of time steps per chunk
        spatial_chunks (dict): Spatial chunking, e.g., {'lat': 50, 'lon': 50}
        """
        if len(file_list) == 0:
            raise ValueError("No files found matching the pattern")
        
        # Set up chunking strategy
        chunks = {'time': time_chunk_size}
        if spatial_chunks:
            chunks.update(spatial_chunks)
            
        print(f"Optimized loading of {len(file_list)} files with chunks: {chunks}")
        
        # Use context manager for better memory management
        with xr.set_options(file_cache_maxsize=128):  # Increase file cache
            dataset = xr.open_mfdataset(
                file_list,
                preprocess=preprocess_func,
                concat_dim='time',
                combine='nested',
                decode_times=False,
                data_vars='minimal',
                coords='minimal',
                parallel=True,
                engine='netcdf4',
                chunks=chunks,
                # Additional optimizations
                drop_variables=None,  # Specify variables to drop if known
                compat='override',    # Speed up coordinate alignment
                join='override'       # Speed up coordinate joining
            )
        
        return dataset
        
    def _get_variables(self, dataset):
        """Extract variable names from a dataset (excluding coordinates)"""
        # Filter out coordinate variables - adjust this based on your data structure
        coords = set(dataset.coords.keys())
        all_vars = set(dataset.data_vars.keys())
        return sorted(list(all_vars - coords))

    
    def compute_coupling_strength(self, var1_data, var2_data, season):
        """
        Compute coupling strength between two variables
        
        Parameters:
        var1_data, var2_data (xarray.DataArray): Input variables
        
        Returns:
        xarray.DataArray: Coupling strength
        """
        xTest = var1_data.copy(deep=True)
        yTest = var2_data.copy(deep=True)
        
        xday = xTest.groupby('time.season')
        yday = yTest.groupby('time.season')
    
        # Get the covariance of the two (numerator in coupling index)
        corrTerm = ((xday - xday.mean()) * (yday - yday.mean())).groupby('time.season').sum() / np.sqrt( ((xday - xday.mean())**2).groupby('time.season').sum()  * ((yday - yday.mean())**2).groupby('time.season').sum() )
        # Now compute the actual coupling index 
        couplingIndex = (corrTerm * yday.std() ).sel(season=season).persist()

        
        return couplingIndex

     
    def create_interactive_plot(self):
        """Create the interactive plotting interface with manual plot button"""

        # Create file type dropdowns
        file_type1_dropdown = widgets.Dropdown(
            options=['lnd', 'atm'],
            value='lnd',
            description='File Type 1:',
            style={'description_width': 'initial'}
        )

        file_type2_dropdown = widgets.Dropdown(
            options=['lnd', 'atm'],
            value='atm',
            description='File Type 2:',
            style={'description_width': 'initial'}
        )
        
        # Create variable dropdowns - will be updated based on file type selection
        var1_dropdown = widgets.Dropdown(
            options=self.variables['lnd'],
            value=self.variables['lnd'][0] if self.variables['lnd'] else None,
            description='Variable 1:',
            style={'description_width': 'initial'}
        )
        
        var2_dropdown = widgets.Dropdown(
            options=self.variables['atm'],
            value=self.variables['atm'][0] if self.variables['atm'] else None,
            description='Variable 2:',
            style={'description_width': 'initial'}
        )
        
        season_dropdown = widgets.Dropdown(
            options=self.seasons,
            value='JJA',
            description='Season:',
            style={'description_width': 'initial'}
        )
        
        # Create the plot button
        plot_button = widgets.Button(
            description='Generate Plot',
            button_style='info',  # 'success', 'info', 'warning', 'danger' or ''
            tooltip='Click to generate/update the plot',
            icon='line-chart'  # FontAwesome icon name (without fa- prefix)
        )
        
        # Create output widget for plots
        output = widgets.Output()
        
        # Function to update variable dropdowns when file type changes
        def update_var1_options(change):
            file_type = change['new']
            var1_dropdown.options = self.variables[file_type]
            if self.variables[file_type]:
                var1_dropdown.value = self.variables[file_type][0]
        
        def update_var2_options(change):
            file_type = change['new']
            var2_dropdown.options = self.variables[file_type]
            if self.variables[file_type]:
                var2_dropdown.value = self.variables[file_type][0]
        
        # Observe changes in file type dropdowns
        file_type1_dropdown.observe(update_var1_options, names='value')
        file_type2_dropdown.observe(update_var2_options, names='value')
        
        # Create the plot generation function
        def generate_plots(button):
            with output:
                output.clear_output(wait=True)  # Clear previous output
                
                try:
                    # Get current values from widgets
                    file_type1 = file_type1_dropdown.value
                    var1 = var1_dropdown.value
                    file_type2 = file_type2_dropdown.value
                    var2 = var2_dropdown.value
                    season = season_dropdown.value

                    landMask = self.datasets['lnd']['landmask']
                    
                    print(f"Generating plot for {var1} ({file_type1}) vs {var2} ({file_type2}) - {season} season...")
                    
                    # Get seasonal data for both variables
                    var1_data = self.datasets[file_type1][var1]
                    var2_data = self.datasets[file_type2][var2]
                    if ((file_type1=='lnd') & (file_type2=='atm')): 
                        var1_data['time'] = var2_data['time']
                    elif ((file_type1=='atm') & (file_type2=='lnd')): 
                        var2_data['time'] = var1_data['time']
                    
                    # Compute coupling strength
                    coupling = self.compute_coupling_strength(var1_data, var2_data, season)
                    
                    # Create subplots
                    # fig,axes = plt.subplots(1,2,figsize=(15,8), subplot_kw={'projection':ccrs.PlateCarree()})
                    # axs = axes.ravel()
                    fig = plt.figure(figsize=(15, 12))
                    # Create custom subplot layout
                    ax1 = plt.subplot(2, 2, 1, projection=ccrs.PlateCarree())  # Top left
                    ax2 = plt.subplot(2, 2, 2, projection=ccrs.PlateCarree())  # Top right
                    ax3 = plt.subplot(2, 1, 2, projection=ccrs.PlateCarree())  # Bottom spanning both columns

                    
                    
                    # Plot Variable 1
                    var1_std_season  = var1_data.groupby('time.season').std().sel(season=season) * landMask
                    longrid, latgrid = np.meshgrid(var1_data['lon'], var1_data['lat'] )
                    norm_sigma_var1  =  matplotlib.colors.Normalize(vmin=np.nanpercentile(var1_std_season, 5),
                                                                    vmax=np.nanpercentile(var1_std_season, 95))
                    i=ax1.pcolormesh(longrid, latgrid, var1_std_season,transform=ccrs.PlateCarree(), 
                             norm=norm_sigma_var1, cmap='viridis')
                    ax1.set_title(f'Std({var1}) [{season}]', fontsize=14)
                    ax1.coastlines()
                    ax1.add_feature(cfeature.BORDERS)
                    ax1.add_feature(states_provinces,edgecolor='k',linewidth=0.5) 
                    # Add Colorbar
                    cb1 = fig.colorbar(i, ax=ax1, shrink=0.6, label=r'$\sigma$ ['+str(var1_data.units)+']', orientation='horizontal', pad=0.025) 
                    
                    # Plot Variable 2
                    var2_std_season  = var2_data.groupby('time.season').std().sel(season=season) * landMask
                    longrid, latgrid = np.meshgrid(var2_data['lon'], var2_data['lat'] )
                    norm_sigma_var2  =  matplotlib.colors.Normalize(vmin=np.nanpercentile(var2_std_season, 5),
                                                                    vmax=np.nanpercentile(var2_std_season, 95))
                    i=ax2.pcolormesh(longrid, latgrid, var2_std_season, transform=ccrs.PlateCarree(), 
                             norm=norm_sigma_var2, cmap='plasma')
                    ax2.set_title(f'Std({var2}) [{season}]', fontsize=14)
                    ax2.coastlines()
                    ax2.add_feature(cfeature.BORDERS)
                    ax2.add_feature(states_provinces,edgecolor='k',linewidth=0.5) 
                    # Add Colorbar
                    cb2 = fig.colorbar(i, ax=ax2, shrink=0.6, label=r'$\sigma$['+str(var2_data.units)+']', orientation='horizontal', pad=0.025) 

                    ## Plot coupling index
                    minCoupling = np.nanpercentile(coupling, 5)
                    maxCoupling = np.nanpercentile(coupling, 95)
                    if abs(maxCoupling)>abs(minCoupling): 
                        diffVal = maxCoupling 
                    elif abs(maxCoupling)<=abs(minCoupling):
                        diffVal = minCoupling
                    norm_coupling  =  matplotlib.colors.Normalize(vmin=-diffVal,vmax=diffVal)
                    i=ax3.pcolormesh(longrid, latgrid, coupling, transform=ccrs.PlateCarree(), 
                             norm=norm_coupling, cmap='Spectral')
                    ax3.set_title(f'Coupling Index ({var1}, {var2})]', fontsize=14)
                    ax3.coastlines()
                    ax3.add_feature(cfeature.BORDERS)
                    ax3.add_feature(states_provinces,edgecolor='k',linewidth=0.5) 
                    # Add Colorbar
                    cb2 = fig.colorbar(i, ax=ax3, shrink=0.6, label='Coupling Index', orientation='horizontal', pad=0.025) 

                    
                    plt.tight_layout()
                    plt.show()
                    
                    # Print coupling strength value
                    print(f"Coupling Strength between {var1} and {var2} for {season}: {coupling}")
                    
                except Exception as e:
                    print(f"Error in plotting: {str(e)}")
                    print("Please check your variable names and data structure.")
        
        # Connect button to plot function
        plot_button.on_click(generate_plots)
        
        # Arrange widgets in a nice layout
        controls_row1 = widgets.HBox([file_type1_dropdown, var1_dropdown])
        controls_row2 = widgets.HBox([file_type2_dropdown, var2_dropdown])
        controls_row3 = widgets.HBox([season_dropdown, plot_button])
        
        controls = widgets.VBox([
            widgets.HTML("<h3>Variable Selection</h3>"),
            controls_row1,
            controls_row2,
            controls_row3,
            widgets.HTML("<hr>")
        ])
        
        # Create the complete interface
        interface = widgets.VBox([controls, output])
        
        return interface

# Usage instructions and example
def main():
    """
    Main function to demonstrate usage
    """
    print("NetCDF Interactive Analysis Tool - Multi-File Land/Atmosphere Version")
    print("=" * 70)
    print("Usage:")
    print("1. Set your data directory and case name pattern")
    print("2. Optional: Define preprocessing functions for land and atmosphere data")
    print("3. Modify the coupling strength formula in compute_coupling_strength()")
    print("4. Run the cells below to start the interactive analysis") 
    print("\nExample:")
    print("# Optional preprocessing functions")
    print("def preprocess_lnd(ds):")
    print("    # Your land data preprocessing here")
    print("    return ds")
    print("")
    print("def preprocess_atm(ds):")
    print("    # Your atmosphere data preprocessing here") 
    print("    return ds")
    print("")
    print("# Initialize analyzer")
    print("analyzer = NetCDFAnalyzer('/path/to/data/', 'case_name', preprocess_lnd, preprocess_atm)")
    print("interactive_widget = analyzer.create_interactive_plot()")
    print("display(interactive_widget)")
    print("\nFeatures:")
    print("- Automatic multi-file loading using glob patterns")
    print("- Support for monthly netCDF files") 
    print("- Optional preprocessing functions for each domain")
    print("- Cross-domain coupling analysis between land and atmosphere")
    print("- Parallel loading for better performance")
    print("- Manual plot generation with button control")


if __name__ == "__main__":
    main()
```

```python
class NetCDFAnalyzer_V2:
    def __init__(self, dataDir, caseNames, preprocess_lnd=None, preprocess_atm=None,
                loading_method='optimized', time_chunks=12, spatial_chunks=None):
        """
        Initialize the analyzer with land and atmosphere netCDF files
        
        Parameters:
        lnd_filepath (str): Path to your land netCDF file
        atm_filepath (str): Path to your atmosphere netCDF file
        """
        self.dataDir = dataDir
        self.caseNames = caseNames

        # Generate file lists
        self.listFiles_lnd = np.sort(glob.glob(dataDir + caseNames + '/lnd/hist/*clm2.h1.????-*'))
        self.listFiles_atm = np.sort(glob.glob(dataDir + caseNames + '/atm/hist/*cam.h1.????-*'))
        
        print(f"Found {len(self.listFiles_lnd)} land files")
        print(f"Found {len(self.listFiles_atm)} atmosphere files")

        # Choose loading method
        if loading_method == 'lazy':
            load_func = self._load_dataset_lazy
        elif loading_method == 'optimized':
            load_func = lambda files, preproc: self._load_dataset_optimized(
                files, preproc, time_chunks, spatial_chunks)
        else:
            load_func = self._load_dataset
        
        # Load datasets using open_mfdataset
        print("Loading land dataset...")
        start_time = pd.Timestamp.now()
        self.lnd_ds = load_func(self.listFiles_lnd, preprocess_lnd)
        lnd_time = pd.Timestamp.now() - start_time
        print(f"Land dataset loaded in {lnd_time.total_seconds():.1f} seconds")
        
        print("Loading atmosphere dataset...")
        start_time = pd.Timestamp.now()
        self.atm_ds = load_func(self.listFiles_atm, preprocess_atm)
        atm_time = pd.Timestamp.now() - start_time
        print(f"Atmosphere dataset loaded in {atm_time.total_seconds():.1f} seconds")

        # Annoyingly, lnd and atm grids aren't always the same, 
        # Lat can differ on order of 1e-7, so just reassign... 
        self.lnd_ds['lat'] = self.atm_ds['lat']
        
        # Store dataset references
        self.datasets = {'lnd': self.lnd_ds, 'atm': self.atm_ds}
        
        # Get variables from both files
        self.lnd_variables = self._get_variables(self.lnd_ds)
        self.atm_variables = self._get_variables(self.atm_ds)
        
        # Create combined variable dictionary
        self.variables = {
            'lnd': self.lnd_variables,
            'atm': self.atm_variables
        }
        
        self.seasons = ['DJF', 'MAM', 'JJA', 'SON']  # Standard meteorological seasons
        
        # Store last computed results for saving
        self.last_results = None

        print("Initialization complete!")
        print(f"Land variables: {len(self.lnd_variables)}")
        print(f"Atmosphere variables: {len(self.atm_variables)}")
        if loading_method == 'lazy':
            print("Note: Data is lazily loaded. Computation will happen when plots are generated.")
    
    
    def _load_dataset(self, file_list, preprocess_func=None):
        """
        Load multiple netCDF files into a single dataset
        
        Parameters:
        file_list (list): List of file paths
        preprocess_func (function, optional): Preprocessing function
        
        Returns:
        xarray.Dataset: Combined dataset
        """
        if len(file_list) == 0:
            raise ValueError("No files found matching the pattern")
        
        print(f"Loading {len(file_list)} files...")
        
        # Optimization 1: Use chunks and engine specifications
        dataset = xr.open_mfdataset(
            file_list,
            preprocess=preprocess_func,
            concat_dim='time',
            combine='nested',
            decode_times=False,
            data_vars='minimal',
            coords='minimal',  # Only load minimal coordinates
            parallel=True,
            engine='netcdf4',  # Specify engine explicitly
            # chunks={'time': 12},  # Chunk by time (adjust based on your data)
            # Alternative chunking strategies:
            chunks={'time': 'auto'}  # Let xarray decide
            # chunks={'time': 6, 'lat': 50, 'lon': 50}  # Multi-dimensional chunking
        )
        
        return dataset
        
    def _load_dataset_lazy(self, file_list, preprocess_func=None):
        """
        Alternative lazy loading approach - only loads metadata initially
        Call .load() or .compute() when you actually need the data
        """
        if len(file_list) == 0:
            raise ValueError("No files found matching the pattern")
        
        print(f"Lazy loading {len(file_list)} files (metadata only)...")
        
        dataset = xr.open_mfdataset(
            file_list,
            preprocess=preprocess_func,
            concat_dim='time',
            combine='nested',
            decode_times=False,
            data_vars='minimal',
            coords='minimal',
            parallel=True,
            engine='netcdf4',
            compat='override',
            chunks={'time': 'auto'},  # Let dask handle chunking
        )
        
        return dataset
    
    def _load_dataset_optimized(self, file_list, preprocess_func=None, 
                               time_chunk_size=12, spatial_chunks=None):
        """
        Highly optimized loading with custom chunk sizes and caching
        
        Parameters:
        file_list (list): List of file paths
        preprocess_func (function, optional): Preprocessing function
        time_chunk_size (int): Number of time steps per chunk
        spatial_chunks (dict): Spatial chunking, e.g., {'lat': 50, 'lon': 50}
        """
        if len(file_list) == 0:
            raise ValueError("No files found matching the pattern")
        
        # Set up chunking strategy
        chunks = {'time': time_chunk_size}
        if spatial_chunks:
            chunks.update(spatial_chunks)
            
        print(f"Optimized loading of {len(file_list)} files with chunks: {chunks}")
        
        # Use context manager for better memory management
        with xr.set_options(file_cache_maxsize=128):  # Increase file cache
            dataset = xr.open_mfdataset(
                file_list,
                preprocess=preprocess_func,
                concat_dim='time',
                combine='nested',
                decode_times=False,
                data_vars='minimal',
                coords='minimal',
                parallel=True,
                engine='netcdf4',
                chunks=chunks,
                # Additional optimizations
                drop_variables=None,  # Specify variables to drop if known
                compat='override',    # Speed up coordinate alignment
                join='override'       # Speed up coordinate joining
            )
        
        return dataset
        
    def _get_variables(self, dataset):
        """Extract variable names from a dataset (excluding coordinates)"""
        # Filter out coordinate variables - adjust this based on your data structure
        coords = set(dataset.coords.keys())
        all_vars = set(dataset.data_vars.keys())
        return sorted(list(all_vars - coords))

    
    def compute_coupling_strength(self, var1_data, var2_data, season):
        """
        Compute coupling strength between two variables
        
        Parameters:
        var1_data, var2_data (xarray.DataArray): Input variables
        
        Returns:
        xarray.DataArray: Coupling strength
        """
        xTest = var1_data.copy(deep=True)
        yTest = var2_data.copy(deep=True)
        
        xday = xTest.groupby('time.season')
        yday = yTest.groupby('time.season')
    
        # Get the covariance of the two (numerator in coupling index)
        corrTerm = ((xday - xday.mean()) * (yday - yday.mean())).groupby('time.season').sum() / np.sqrt( ((xday - xday.mean())**2).groupby('time.season').sum()  * ((yday - yday.mean())**2).groupby('time.season').sum() )
        # Now compute the actual coupling index 
        couplingIndex = (corrTerm * yday.std() ).sel(season=season).persist()

        
        return couplingIndex

    def save_results_to_netcdf(self, var1_std, var2_std, coupling, 
                               var1_name, var2_name, season, output_path=None):
        """
        Save the three computed variables to a netCDF file
        
        Parameters:
        var1_std (xarray.DataArray): Standard deviation of variable 1
        var2_std (xarray.DataArray): Standard deviation of variable 2
        coupling (xarray.DataArray): Coupling index
        var1_name (str): Name of variable 1
        var2_name (str): Name of variable 2
        season (str): Season identifier
        output_path (str, optional): Path for output file. If None, auto-generates name
        """
        if output_path is None:
            output_path = f"{self.caseNames}_coupling_{var1_name}_{var2_name}_{season}.nc"
        
        # Create a new dataset with the three variables
        ds_out = xr.Dataset({
            f'{var1_name}_std': var1_std,
            f'{var2_name}_std': var2_std,
            f'coupling_{var1_name}_{var2_name}': coupling
        })
        
        # Add metadata attributes
        ds_out.attrs['title'] = f'Coupling Analysis: {var1_name} and {var2_name}'
        ds_out.attrs['season'] = season
        ds_out.attrs['case_name'] = self.caseNames
        ds_out.attrs['creation_date'] = pd.Timestamp.now().strftime('%Y-%m-%d %H:%M:%S')
        
        # Add variable-specific attributes
        ds_out[f'{var1_name}_std'].attrs['long_name'] = f'Standard deviation of {var1_name}'
        ds_out[f'{var1_name}_std'].attrs['season'] = season
        
        ds_out[f'{var2_name}_std'].attrs['long_name'] = f'Standard deviation of {var2_name}'
        ds_out[f'{var2_name}_std'].attrs['season'] = season
        
        ds_out[f'coupling_{var1_name}_{var2_name}'].attrs['long_name'] = f'Coupling index between {var1_name} and {var2_name}'
        ds_out[f'coupling_{var1_name}_{var2_name}'].attrs['season'] = season
        
        # Save to netCDF
        print(f"Saving results to: {output_path}")
        ds_out.to_netcdf(output_path)
        print(f"Successfully saved to {output_path}")
        
        return output_path

     
    def create_interactive_plot(self):
        """Create the interactive plotting interface with manual plot button and save option"""

        # Create file type dropdowns
        file_type1_dropdown = widgets.Dropdown(
            options=['lnd', 'atm'],
            value='lnd',
            description='File Type 1:',
            style={'description_width': 'initial'}
        )

        file_type2_dropdown = widgets.Dropdown(
            options=['lnd', 'atm'],
            value='atm',
            description='File Type 2:',
            style={'description_width': 'initial'}
        )
        
        # Create variable dropdowns - will be updated based on file type selection
        var1_dropdown = widgets.Dropdown(
            options=self.variables['lnd'],
            value=self.variables['lnd'][0] if self.variables['lnd'] else None,
            description='Variable 1:',
            style={'description_width': 'initial'}
        )
        
        var2_dropdown = widgets.Dropdown(
            options=self.variables['atm'],
            value=self.variables['atm'][0] if self.variables['atm'] else None,
            description='Variable 2:',
            style={'description_width': 'initial'}
        )
        
        season_dropdown = widgets.Dropdown(
            options=self.seasons,
            value='JJA',
            description='Season:',
            style={'description_width': 'initial'}
        )
        
        # Create the plot button
        plot_button = widgets.Button(
            description='Generate Plot',
            button_style='info',
            tooltip='Click to generate/update the plot',
            icon='line-chart'
        )
        
        # Create save controls
        save_checkbox = widgets.Checkbox(
            value=False,
            description='Save results to netCDF',
            style={'description_width': 'initial'}
        )
        
        output_filename = widgets.Text(
            value='',
            placeholder='Leave blank for auto-generated name',
            description='Output filename:',
            style={'description_width': 'initial'},
            layout=widgets.Layout(width='400px')
        )
        
        save_button = widgets.Button(
            description='Save to NetCDF',
            button_style='success',
            tooltip='Save the last plotted results',
            icon='save',
            disabled=True  # Initially disabled until results are generated
        )
        
        # Create output widget for plots
        output = widgets.Output()
        
        # Function to update variable dropdowns when file type changes
        def update_var1_options(change):
            file_type = change['new']
            var1_dropdown.options = self.variables[file_type]
            if self.variables[file_type]:
                var1_dropdown.value = self.variables[file_type][0]
        
        def update_var2_options(change):
            file_type = change['new']
            var2_dropdown.options = self.variables[file_type]
            if self.variables[file_type]:
                var2_dropdown.value = self.variables[file_type][0]
        
        # Observe changes in file type dropdowns
        file_type1_dropdown.observe(update_var1_options, names='value')
        file_type2_dropdown.observe(update_var2_options, names='value')
        
        # Create the plot generation function
        def generate_plots(button):
            with output:
                output.clear_output(wait=True)
                
                try:
                    # Get current values from widgets
                    file_type1 = file_type1_dropdown.value
                    var1 = var1_dropdown.value
                    file_type2 = file_type2_dropdown.value
                    var2 = var2_dropdown.value
                    season = season_dropdown.value

                    landMask = self.datasets['lnd']['landmask']
                    
                    print(f"Generating plot for {var1} ({file_type1}) vs {var2} ({file_type2}) - {season} season...")
                    
                    # Get seasonal data for both variables
                    var1_data = self.datasets[file_type1][var1]
                    var2_data = self.datasets[file_type2][var2]
                    if ((file_type1=='lnd') & (file_type2=='atm')): 
                        var1_data['time'] = var2_data['time']
                    elif ((file_type1=='atm') & (file_type2=='lnd')): 
                        var2_data['time'] = var1_data['time']
                    
                    # Compute coupling strength
                    coupling = self.compute_coupling_strength(var1_data, var2_data, season)
                    
                    # Compute standard deviations
                    var1_std_season = var1_data.groupby('time.season').std().sel(season=season) * landMask
                    var2_std_season = var2_data.groupby('time.season').std().sel(season=season) * landMask
                    
                    # Store results for potential saving
                    self.last_results = {
                        'var1_std': var1_std_season,
                        'var2_std': var2_std_season,
                        'coupling': coupling,
                        'var1_name': var1,
                        'var2_name': var2,
                        'season': season
                    }
                    
                    # Enable save button now that we have results
                    save_button.disabled = False
                    
                    # Create subplots
                    fig = plt.figure(figsize=(15, 12))
                    ax1 = plt.subplot(2, 2, 1, projection=ccrs.PlateCarree())
                    ax2 = plt.subplot(2, 2, 2, projection=ccrs.PlateCarree())
                    ax3 = plt.subplot(2, 1, 2, projection=ccrs.PlateCarree())
                    
                    # Plot Variable 1
                    longrid, latgrid = np.meshgrid(var1_data['lon'], var1_data['lat'])
                    norm_sigma_var1 = matplotlib.colors.Normalize(
                        vmin=np.nanpercentile(var1_std_season, 5),
                        vmax=np.nanpercentile(var1_std_season, 95)
                    )
                    i = ax1.pcolormesh(longrid, latgrid, var1_std_season,
                                      transform=ccrs.PlateCarree(), 
                                      norm=norm_sigma_var1, cmap='viridis')
                    ax1.set_title(f'Std({var1}) [{season}]', fontsize=14)
                    ax1.coastlines()
                    ax1.add_feature(cfeature.BORDERS)
                    ax1.add_feature(states_provinces, edgecolor='k', linewidth=0.5)
                    cb1 = fig.colorbar(i, ax=ax1, shrink=0.6, 
                                      label=r'$\sigma$ ['+str(var1_data.units)+']', 
                                      orientation='horizontal', pad=0.025)
                    
                    # Plot Variable 2
                    longrid, latgrid = np.meshgrid(var2_data['lon'], var2_data['lat'])
                    norm_sigma_var2 = matplotlib.colors.Normalize(
                        vmin=np.nanpercentile(var2_std_season, 5),
                        vmax=np.nanpercentile(var2_std_season, 95)
                    )
                    i = ax2.pcolormesh(longrid, latgrid, var2_std_season,
                                      transform=ccrs.PlateCarree(), 
                                      norm=norm_sigma_var2, cmap='plasma')
                    ax2.set_title(f'Std({var2}) [{season}]', fontsize=14)
                    ax2.coastlines()
                    ax2.add_feature(cfeature.BORDERS)
                    ax2.add_feature(states_provinces, edgecolor='k', linewidth=0.5)
                    cb2 = fig.colorbar(i, ax=ax2, shrink=0.6, 
                                      label=r'$\sigma$['+str(var2_data.units)+']', 
                                      orientation='horizontal', pad=0.025)

                    # Plot coupling index
                    minCoupling = np.nanpercentile(coupling, 5)
                    maxCoupling = np.nanpercentile(coupling, 95)
                    if abs(maxCoupling) > abs(minCoupling): 
                        diffVal = maxCoupling 
                    elif abs(maxCoupling) <= abs(minCoupling):
                        diffVal = minCoupling
                    norm_coupling = matplotlib.colors.Normalize(vmin=-diffVal, vmax=diffVal)
                    i = ax3.pcolormesh(longrid, latgrid, coupling,
                                      transform=ccrs.PlateCarree(), 
                                      norm=norm_coupling, cmap='Spectral')
                    ax3.set_title(f'Coupling Index ({var1}, {var2})]', fontsize=14)
                    ax3.coastlines()
                    ax3.add_feature(cfeature.BORDERS)
                    ax3.add_feature(states_provinces, edgecolor='k', linewidth=0.5)
                    cb2 = fig.colorbar(i, ax=ax3, shrink=0.6, label='Coupling Index', 
                                      orientation='horizontal', pad=0.025)
                    
                    plt.tight_layout()
                    plt.show()
                    
                    print(f"Coupling Strength between {var1} and {var2} for {season}: {coupling}")
                    print("\nResults stored. Click 'Save to NetCDF' button to export.")
                    
                except Exception as e:
                    print(f"Error in plotting: {str(e)}")
                    print("Please check your variable names and data structure.")
        
        # Create the save function
        def save_results(button):
            with output:
                if self.last_results is None:
                    print("No results to save. Please generate a plot first.")
                    return
                
                try:
                    output_path = output_filename.value if output_filename.value else None
                    
                    saved_file = self.save_results_to_netcdf(
                        self.last_results['var1_std'],
                        self.last_results['var2_std'],
                        self.last_results['coupling'],
                        self.last_results['var1_name'],
                        self.last_results['var2_name'],
                        self.last_results['season'],
                        output_path
                    )
                    
                except Exception as e:
                    print(f"Error saving file: {str(e)}")
        
        # Connect buttons to functions
        plot_button.on_click(generate_plots)
        save_button.on_click(save_results)
        
        # Arrange widgets in a nice layout
        controls_row1 = widgets.HBox([file_type1_dropdown, var1_dropdown])
        controls_row2 = widgets.HBox([file_type2_dropdown, var2_dropdown])
        controls_row3 = widgets.HBox([season_dropdown, plot_button])
        
        save_controls = widgets.VBox([
            widgets.HTML("<h4>Save Options</h4>"),
            save_checkbox,
            output_filename,
            save_button
        ])
        
        controls = widgets.VBox([
            widgets.HTML("<h3>Variable Selection</h3>"),
            controls_row1,
            controls_row2,
            controls_row3,
            widgets.HTML("<hr>"),
            save_controls,
            widgets.HTML("<hr>")
        ])
        
        # Create the complete interface
        interface = widgets.VBox([controls, output])
        
        return interface


# Usage instructions and example
def main():
    """
    Main function to demonstrate usage
    """
    print("NetCDF Interactive Analysis Tool - Multi-File Land/Atmosphere Version")
    print("=" * 70)
    print("Usage:")
    print("1. Set your data directory and case name pattern")
    print("2. Optional: Define preprocessing functions for land and atmosphere data")
    print("3. Modify the coupling strength formula in compute_coupling_strength()")
    print("4. Run the cells below to start the interactive analysis") 
    print("\nExample:")
    print("# Optional preprocessing functions")
    print("def preprocess_lnd(ds):")
    print("    # Your land data preprocessing here")
    print("    return ds")
    print("")
    print("def preprocess_atm(ds):")
    print("    # Your atmosphere data preprocessing here") 
    print("    return ds")
    print("")
    print("# Initialize analyzer")
    print("analyzer = NetCDFAnalyzer('/path/to/data/', 'case_name', preprocess_lnd, preprocess_atm)")
    print("interactive_widget = analyzer.create_interactive_plot()")
    print("display(interactive_widget)")
    print("\nFeatures:")
    print("- Automatic multi-file loading using glob patterns")
    print("- Support for monthly netCDF files") 
    print("- Optional preprocessing functions for each domain")
    print("- Cross-domain coupling analysis between land and atmosphere")
    print("- Parallel loading for better performance")
    print("- Manual plot generation with button control")
    print("- Save computed results (std and coupling index) to netCDF")


if __name__ == "__main__":
    main()
```
