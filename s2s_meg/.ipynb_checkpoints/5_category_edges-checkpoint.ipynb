# Generate category edges of all zarrs for use in the RPS metric in climpred

---

Creates category edges for the anomaly and verification files (both daily and biweekly will work). The verification calculation happens at the end as it is different than the models.

```python
import cftime
import numpy as np
import xarray as xr
xr.set_options(keep_attrs=True)
import climpred
from tqdm import tqdm
import dask.array as da
import matplotlib.pyplot as plt
from matplotlib.ticker import FixedLocator
import xskillscore as xs
import warnings
warnings.filterwarnings("ignore")

from dask.distributed import Client
import dask.config
dask.config.set({"array.slicing.split_large_chunks": False})
```

```python
client = Client("tcp://128.117.208.100:35551")
```

Choose your model and time period

```python
model = "CESM2_orig" #ECMWF, ECCC, or NCEP
time = "biweekly" #biweekly or daily
```

```python
hinda2 = xr.open_zarr("/glade/campaign/mmm/c3we/jaye/S2S_zarr/"+model+".anom."+time+".geospatial.zarr/", consolidated=True).astype('float32')
```

```python
hinda2
```

```python
groupby_time = 'dayofyear'
```

```python
hinda2['lead'].attrs['units'] = 'days'
```

```python
hinda2 = hinda2["psl"].chunk({"member": -1, "init": -1, "lead": "auto", "lat": 45, "lon": 60}).persist()
he = climpred.HindcastEnsemble(hinda2)
```

Calculate the edges then write out to `zarr`

```python
#hinda2_edges = he.get_initialized().groupby(f'init.{groupby_time}').quantile(q=[1./5.,2./5.,3./5.,4./5.], dim=['init','member'], skipna=False).rename({'quantile':'category_edge'}).astype('float32').persist()
hinda2_edges = he.get_initialized().groupby(f'init.{groupby_time}').quantile(q=[1./3.,2./3.], dim=['init','member'], skipna=False).rename({'quantile':'category_edge'}).astype('float32').persist()
```

```python
hinda2_edges
```

```python
hinda2_edges = hinda2_edges.chunk({"category_edge": "auto", "dayofyear": "auto", "lead": "auto", "lat": 45, "lon": 60}).persist()
```

```python
#hinda2_edges["t2m"].isel({groupby_time:[1,7]}).plot(row='category_edge', lead=2,col=groupby_time, robust=True)
```

```python
hinda2_edges
```

```python
%time hinda2_edges.to_zarr("/glade/campaign/mmm/c3we/jaye/S2S_zarr/"+model+".anom.cat_edges."+time+".geospatial.zarr/",mode="a",consolidated=True)
#%time hinda2_edges.to_zarr("/glade/campaign/mmm/c3we/jaye/S2S_zarr/"+model+".anom.cat_edges_quint."+time+".geospatial.zarr/",mode="w",consolidated=True)
```

# Now do the same for verification!

We also make a plot to make sure it looks right

```python
verif = xr.open_zarr("/glade/campaign/mmm/c3we/jaye/S2S_zarr/OBS_SUBX.anom."+time+".geospatial.zarr/", consolidated=True).astype('float32')
```

```python
groupby_time = 'dayofyear'
```

```python
%time verif_edges = verif.chunk({'time':-1,'lon':'auto'}).groupby(f'time.{groupby_time}').quantile(q=[1./3.,2./3.], dim='time', skipna=False).rename({'quantile':'category_edge'}).astype('float32').compute()
```

```python
verif_edges["t2m"].isel({groupby_time:[1,7]}).plot(row='category_edge', col=groupby_time, robust=True)
```

```python
verif_edges
```

```python
%time verif_edges.to_zarr("/glade/campaign/mmm/c3we/jaye/S2S_zarr/OBS_SUBX.anom.cat_edges."+time+".geospatial.zarr/",mode="w",consolidated=True)
```
