```python
def regrid_data(fromthis, tothis, method=1):
    """Regrid data using various different methods"""

    #Import necessary modules:
    import xarray as xr

    if method == 1:
        # kludgy: spatial regridding only, seems like can't automatically deal with time
        if 'time' in fromthis.coords:
            result = [fromthis.isel(time=t).interp_like(tothis) for t,time in enumerate(fromthis['time'])]
            result = xr.concat(result, 'time')
            return result
        else:
            return fromthis.interp_like(tothis)
    elif method == 2:
        newlat = tothis['lat']
        newlon = tothis['lon']
        coords = dict(fromthis.coords)
        coords['lat'] = newlat
        coords['lon'] = newlon
        return fromthis.interp(coords)
    elif method == 3:
        newlat = tothis['lat']
        newlon = tothis['lon']
        ds_out = xr.Dataset({'lat': newlat, 'lon': newlon})
        regridder = xe.Regridder(fromthis, ds_out, 'bilinear')
        return regridder(fromthis)
    elif method==4:
        # geocat
        newlat = tothis['lat']
        newlon = tothis['lon']
        result = geocat.comp.linint2(fromthis, newlon, newlat, False)
        result.name = fromthis.name
        
        return result

def update_leads(x):
    """Converts from time coordinates to lead coordinates.
    I.e., lead days.
    """
    x = x.rename({"time": "lead"})
    x = x.assign_coords(lead=np.arange(x.lead.size).astype(int))
    
    return x

def drop_duplicates_along_all_dims(obj, keep=False):
    deduplicated = obj
    for dim in obj.dims:
        indexes = {dim: ~deduplicated.get_index(dim).duplicated(keep=keep)}
        deduplicated = deduplicated.isel(indexes)
    return deduplicated
```

```python
## CLAUDE: Approach 1: Domain-averaged anomaly with spatial terciles
#      Calculate the spatial mean for each time step, then classify based on the distribution of those means:
#      Interpretation: "Days when the entire region is anomalously dry/wet on average"


## What if I only care about a certain season? 
seasonSel = 'JJA'

if seasonSel=='ANN':
    variable=hotspot_SM
else: 
    variable=hotspot_SM.sel(init=hotspot_SM.init.dt.season == seasonSel)
    
# Calculate spatial mean for each time step (averaging over the valid domain)
domain_mean = variable.mean(dim=['lat', 'lon'], skipna=True)

# Calculate terciles of the domain-mean time series
tercile_33 = domain_mean.chunk(dict(init=-1)).quantile(1/3, dim='init')
tercile_67 = domain_mean.chunk(dict(init=-1)).quantile(2/3, dim='init')

print(f"Domain mean tercile 33rd: {tercile_33.values:.3f}")
print(f"Domain mean tercile 67th: {tercile_67.values:.3f}")

# Classify each day based on its domain mean
categories_all = np.full([len(variable.init.values)], '             ')
for iT, mean_val in enumerate(domain_mean.values):
    if mean_val <= tercile_33:
        categories_all[iT] = 'dry'
    elif mean_val >= tercile_67:
        categories_all[iT] = 'wet'
    else:
        categories_all[iT] = 'neutral'
```

```python
print('For strongly-coupled region: \n %6i dry days\n %6i wet days\n %6i neutral days\n' % 
      (
       len(np.where(categories_all[:]=='dry')[0]),
       len(np.where(categories_all[:]=='wet')[0]), 
       len(np.where(categories_all[:]=='neutral')[0])) 
     )
```

```python
## CLAUDE: Approach 2: Spatial terciles with coherence threshold
#     Use domain-wide terciles (not per-gridpoint) and require spatial agreement:
#     Interpretation: "Days when at least X% of the region shows extreme (dry/wet) conditions relative to the overall distribution"
seasonSel = 'JJA'

if seasonSel=='ANN':
    variable=hotspot_SM
else: 
    variable=hotspot_SM.sel(init=hotspot_SM.init.dt.season == seasonSel)

# Calculate terciles across ENTIRE domain (space + time)
variable_flat = variable.stack(space=('lat', 'lon'))
tercile_33 = variable_flat.chunk(dict(init=-1)).quantile(1/3, dim=['init', 'space'], skipna=True)
tercile_67 = variable_flat.chunk(dict(init=-1)).quantile(2/3, dim=['init', 'space'], skipna=True)

print(f"Domain-wide tercile 33rd: {tercile_33.values:.3f}")
print(f"Domain-wide tercile 67th: {tercile_67.values:.3f}")

# Classify each grid point for each time step
soil_categories = xr.where(
    variable <= tercile_33, 0,  # dry
    xr.where( 
        variable >= tercile_67, 2,  # wet
        1  # neutral
    ))

# Now count spatial coherence
soilCats = soil_categories.values
n_times, n_lat, n_lon = soilCats.shape
domain_reshaped = soilCats.reshape(n_times, n_lat * n_lon)

threshold = 0.5  # or lower, like 0.3-0.4
categories_all = np.full([len(variable.init.values)], '             ')

for iT in range(n_times):
    valid_mask = np.isfinite(domain_reshaped[iT,:])
    valid_data = domain_reshaped[iT, valid_mask]
    nSpace_valid = len(valid_data)
    
    iDry     = np.sum(valid_data == 0)
    iNeutral = np.sum(valid_data == 1)
    iWet     = np.sum(valid_data == 2)
    
    if (iDry >= (nSpace_valid * threshold)):
        categories_all[iT] = 'dry'
    elif (iWet >= (nSpace_valid * threshold)):
        categories_all[iT] = 'wet'
    elif (iNeutral >= (nSpace_valid * threshold)):
        categories_all[iT] = 'neutral'
    else: 
        categories_all[iT] = 'mixed'

print('For strongly-coupled region: \n %6i dry days\n %6i wet days\n %6i neutral days\n' % 
      (
       len(np.where(categories_all[:]=='dry')[0]),
       len(np.where(categories_all[:]=='wet')[0]), 
       len(np.where(categories_all[:]=='neutral')[0])) 
     )
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
import seaborn as sns

# Set style
sns.set_style("whitegrid")
plt.rcParams['figure.figsize'] = (12, 8)

def calculate_spatial_acc(forecast_anom, observed_anom):
    """
    Calculate Anomaly Correlation Coefficient over a spatial domain.
    
    Parameters:
    -----------
    forecast_anom : np.array
        Forecast anomalies (spatial grid flattened or multi-point)
    observed_anom : np.array
        Observed anomalies (same shape as forecast)
    
    Returns:
    --------
    acc : float
        Spatial ACC value
    """
    # Flatten if needed
    fcst = forecast_anom.flatten()
    obs = observed_anom.flatten()
    
    # Remove any NaN values
    mask = ~(np.isnan(fcst) | np.isnan(obs))
    fcst = fcst[mask]
    obs = obs[mask]
    
    if len(fcst) < 2:
        return np.nan
    
    # Calculate correlation
    correlation = np.corrcoef(fcst, obs)[0, 1]
    return correlation


def prepare_sample_data(n_events=100, n_spatial_points=50):
    """
    Generate sample data for demonstration.
    Replace this with your actual data loading function.
    
    Parameters:
    -----------
    n_events : int
        Number of forecast initialization events
    n_spatial_points : int
        Number of spatial points for ACC calculation
    
    Returns:
    --------
    df : pd.DataFrame
        DataFrame with columns: event_id, soil_moisture_anom, spatial_acc, 
        init_date, season (optional)
    """
    np.random.seed(42)
    
    data = []
    for i in range(n_events):
        # Generate soil moisture anomaly (standardized units)
        sm_anom = np.random.randn()
        
        # Simulate spatial ACC that has some dependency on soil moisture
        # Higher absolute SM anomaly -> slightly better ACC (as an example)
        base_acc = 0.4 + 0.15 * np.abs(sm_anom) + np.random.randn() * 0.2
        spatial_acc = np.clip(base_acc, -0.5, 0.95)  # Keep in realistic range
        
        # Add seasonal information
        season = np.random.choice(['DJF', 'MAM', 'JJA', 'SON'])
        
        data.append({
            'event_id': i,
            'soil_moisture_anom': sm_anom,
            'spatial_acc': spatial_acc,
            'season': season,
            'init_date': pd.Timestamp('2020-01-01') + pd.Timedelta(days=i*10)
        })
    
    return pd.DataFrame(data)


def regression_analysis(df, sm_col='soil_moisture_anom', acc_col='spatial_acc',
                       control_vars=None, add_quadratic=False):
    """
    Perform regression analysis of spatial ACC vs soil moisture anomaly.
    
    Parameters:
    -----------
    df : pd.DataFrame
        Data with soil moisture and ACC values
    sm_col : str
        Column name for soil moisture anomaly
    acc_col : str
        Column name for spatial ACC
    control_vars : list
        List of additional control variable column names
    add_quadratic : bool
        Whether to add quadratic term for soil moisture
    
    Returns:
    --------
    results : dict
        Dictionary containing regression results
    """
    # Prepare data
    X = df[[sm_col]].values
    y = df[acc_col].values
    
    # Remove NaN values
    mask = ~(np.isnan(X).any(axis=1) | np.isnan(y))
    X = X[mask]
    y = y[mask]
    
    # Add quadratic term if requested
    if add_quadratic:
        X_squared = X ** 2
        X = np.hstack([X, X_squared])
        feature_names = [sm_col, f'{sm_col}_squared']
    else:
        feature_names = [sm_col]
    
    # Add control variables if provided
    if control_vars:
        for var in control_vars:
            X_control = df.loc[mask, var].values.reshape(-1, 1)
            X = np.hstack([X, X_control])
            feature_names.append(var)
    
    # Fit regression
    model = LinearRegression()
    model.fit(X, y)
    y_pred = model.predict(X)
    
    # Calculate statistics
    r2 = r2_score(y, y_pred)
    n = len(y)
    k = X.shape[1]
    
    # Calculate standard errors and p-values
    residuals = y - y_pred
    mse = np.sum(residuals**2) / (n - k - 1)
    var_coef = mse * np.linalg.inv(X.T @ X).diagonal()
    std_errors = np.sqrt(var_coef)
    t_stats = model.coef_ / std_errors
    p_values = 2 * (1 - stats.t.cdf(np.abs(t_stats), n - k - 1))
    
    # Pearson correlation (for simple case)
    if X.shape[1] == 1:
        pearson_r, pearson_p = stats.pearsonr(X.flatten(), y)
    else:
        pearson_r, pearson_p = None, None
    
    results = {
        'model': model,
        'X': X,
        'y': y,
        'y_pred': y_pred,
        'r2': r2,
        'coefficients': model.coef_,
        'intercept': model.intercept_,
        'std_errors': std_errors,
        'p_values': p_values,
        'feature_names': feature_names,
        'pearson_r': pearson_r,
        'pearson_p': pearson_p,
        'n_events': n
    }
    
    return results


def print_results(results):
    """Print regression results in readable format."""
    print("=" * 60)
    print("REGRESSION RESULTS: Spatial ACC ~ Soil Moisture Anomaly")
    print("=" * 60)
    print(f"\nNumber of events: {results['n_events']}")
    print(f"R = {results['r2']:.4f}")
    if results['pearson_r'] is not None:
        print(f"Pearson r = {results['pearson_r']:.4f} (p = {results['pearson_p']:.4f})")
    
    print(f"\nIntercept: {results['intercept']:.4f}")
    print("\nCoefficients:")
    print("-" * 60)
    print(f"{'Variable':<25} {'Coef':<12} {'Std Err':<12} {'p-value':<12}")
    print("-" * 60)
    
    for i, name in enumerate(results['feature_names']):
        coef = results['coefficients'][i]
        se = results['std_errors'][i]
        pval = results['p_values'][i]
        sig = "***" if pval < 0.001 else "**" if pval < 0.01 else "*" if pval < 0.05 else ""
        print(f"{name:<25} {coef:>11.4f} {se:>11.4f} {pval:>11.4f} {sig}")
    
    print("-" * 60)
    print("\nInterpretation:")
    coef_sm = results['coefficients'][0]
    if coef_sm > 0:
        print(f"   Each 1 increase in soil moisture anomaly is associated with")
        print(f"    a {coef_sm:.4f} increase in spatial ACC at week 3-4")
    else:
        print(f"   Each 1 increase in soil moisture anomaly is associated with")
        print(f"    a {abs(coef_sm):.4f} decrease in spatial ACC at week 3-4")
    print("=" * 60)


def plot_regression_results(df, results, sm_col='soil_moisture_anom', 
                           acc_col='spatial_acc', season_col=None):
    """
    Create visualization of regression results.
    
    Parameters:
    -----------
    df : pd.DataFrame
        Original dataframe
    results : dict
        Results from regression_analysis
    sm_col : str
        Soil moisture column name
    acc_col : str
        ACC column name
    season_col : str
        Optional season column for color coding
    """
    fig, axes = plt.subplots(2, 2, figsize=(14, 10))
    
    # Get data (handle NaN masking)
    mask = ~(np.isnan(df[sm_col]) | np.isnan(df[acc_col]))
    sm_data = df.loc[mask, sm_col].values
    acc_data = df.loc[mask, acc_col].values
    
    # Plot 1: Scatter with regression line
    ax = axes[0, 0]
    if season_col and season_col in df.columns:
        seasons = df.loc[mask, season_col]
        for season in seasons.unique():
            season_mask = seasons == season
            ax.scatter(sm_data[season_mask], acc_data[season_mask], 
                      label=season, alpha=0.6, s=50)
    else:
        ax.scatter(sm_data, acc_data, alpha=0.6, s=50, color='steelblue')
    
    # Add regression line (for simple linear case)
    if results['X'].shape[1] == 1:
        sm_range = np.linspace(sm_data.min(), sm_data.max(), 100)
        acc_pred = results['intercept'] + results['coefficients'][0] * sm_range
        ax.plot(sm_range, acc_pred, 'r-', linewidth=2, label='Regression line')
        
        # Add confidence interval
        from scipy import stats as sp_stats
        predict_err = np.sqrt(np.mean((results['y'] - results['y_pred'])**2))
        ci = 1.96 * predict_err
        ax.fill_between(sm_range, acc_pred - ci, acc_pred + ci, 
                        alpha=0.2, color='red', label='95% CI')
    
    ax.set_xlabel('Soil Moisture Anomaly ()', fontsize=11)
    ax.set_ylabel('Spatial ACC (Week 3-4)', fontsize=11)
    ax.set_title(f'ACC vs Soil Moisture (R = {results["r2"]:.3f})', fontsize=12, fontweight='bold')
    ax.axhline(y=0, color='k', linestyle='--', alpha=0.3)
    ax.axvline(x=0, color='k', linestyle='--', alpha=0.3)
    ax.legend()
    ax.grid(True, alpha=0.3)
    
    # Plot 2: Residuals vs Fitted
    ax = axes[0, 1]
    residuals = results['y'] - results['y_pred']
    ax.scatter(results['y_pred'], residuals, alpha=0.6, s=50, color='steelblue')
    ax.axhline(y=0, color='r', linestyle='--', linewidth=2)
    ax.set_xlabel('Fitted Values', fontsize=11)
    ax.set_ylabel('Residuals', fontsize=11)
    ax.set_title('Residual Plot', fontsize=12, fontweight='bold')
    ax.grid(True, alpha=0.3)
    
    # Plot 3: Distribution of ACC by SM terciles
    ax = axes[1, 0]
    terciles = pd.qcut(sm_data, q=3, labels=['Dry', 'Normal', 'Wet'])
    tercile_df = pd.DataFrame({'tercile': terciles, 'acc': acc_data})
    tercile_df.boxplot(column='acc', by='tercile', ax=ax)
    ax.set_xlabel('Soil Moisture Tercile', fontsize=11)
    ax.set_ylabel('Spatial ACC', fontsize=11)
    ax.set_title('ACC Distribution by SM Tercile', fontsize=12, fontweight='bold')
    plt.sca(ax)
    plt.xticks(rotation=0)
    ax.get_figure().suptitle('')  # Remove automatic title
    
    # Plot 4: Time series or histogram of residuals
    ax = axes[1, 1]
    ax.hist(residuals, bins=20, edgecolor='black', alpha=0.7, color='steelblue')
    ax.axvline(x=0, color='r', linestyle='--', linewidth=2)
    ax.set_xlabel('Residuals', fontsize=11)
    ax.set_ylabel('Frequency', fontsize=11)
    ax.set_title('Residual Distribution', fontsize=12, fontweight='bold')
    ax.grid(True, alpha=0.3, axis='y')
    
    plt.tight_layout()
    return fig


# ============================================================================
# MAIN EXECUTION
# ============================================================================

if __name__ == "__main__":
    # Step 1: Load or generate data
    print("Loading data...")
    df = prepare_sample_data(n_events=100, n_spatial_points=50)
    
    # YOUR DATA LOADING WOULD LOOK LIKE:
    # df = pd.read_csv('your_data.csv')
    # Or build df from your forecast and observation arrays
    
    # Step 2: Run regression analysis
    print("\nRunning regression analysis...")
    
    # Simple linear regression
    results_linear = regression_analysis(df, 
                                        sm_col='soil_moisture_anom',
                                        acc_col='spatial_acc',
                                        add_quadratic=False)
    
    print_results(results_linear)
    
    # Optional: Try with quadratic term
    print("\n\nTrying with quadratic term...")
    results_quadratic = regression_analysis(df,
                                           sm_col='soil_moisture_anom',
                                           acc_col='spatial_acc',
                                           add_quadratic=True)
    
    print_results(results_quadratic)
    
    # Step 3: Visualize results
    print("\nGenerating plots...")
    fig = plot_regression_results(df, results_linear, 
                                 sm_col='soil_moisture_anom',
                                 acc_col='spatial_acc',
                                 season_col='season')
    plt.show()
    
    # Step 4: Additional analysis by season (optional)
    print("\n\nSeasonal breakdown:")
    print("=" * 60)
    for season in df['season'].unique():
        df_season = df[df['season'] == season]
        if len(df_season) > 10:  # Only if enough samples
            results_season = regression_analysis(df_season,
                                                sm_col='soil_moisture_anom',
                                                acc_col='spatial_acc')
            print(f"\n{season}: R = {results_season['r2']:.3f}, "
                  f"Coef = {results_season['coefficients'][0]:.4f}, "
                  f"p = {results_season['p_values'][0]:.4f}")
```
