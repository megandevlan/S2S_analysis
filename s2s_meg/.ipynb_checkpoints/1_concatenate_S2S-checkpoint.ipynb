```python
def preprocess_atm(DS): 
    DSsel = DS['TREFHT']

    DSsel = DSsel.assign_coords({"init": DSsel.time.values[0]})

    DSsel = DSsel.rename({"time": "lead"})
    DSsel = DSsel.assign_coords(lead=np.arange(DSsel.lead.size).astype(int))
    
    return DSsel
    
```

```python
# S2Sensemble = []

# Loop through each initialization and concatenate.
for iEns in tqdm(range(11)):
# for iEns in tqdm(range(1)):
    if iEns<10: 
        ensMemberStr = '0'+str(iEns)
    else: 
        ensMemberStr = str(iEns)
        
    filelist = np.sort( glob.glob(filepath+ "*/atm/hist/cesm2cam6climoOCNclimoATM.????-??-??."+ensMemberStr+".cam.h2.*.nc") )

    memberDS = xr.open_mfdataset(
        filelist,
        combine="nested",
        parallel=True,
        concat_dim="init",
        preprocess=preprocess_atm,
        chunks = {"time": -1, "lon": -1, "lat": -1},
        # speeds things up a bit
        coords="minimal",
        compat="override",
    )

    memberDS = memberDS.assign_coords({"member": iEns})

    if iEns==0:
        S2Sensemble = memberDS
    else: 
        S2Sensemble = xr.concat([S2Sensemble, memberDS], "member")
    
```

```python
S2Sensemble = S2Sensemble.sortby("init")
S2Sensemble['init'] = S2Sensemble.indexes['init'].to_datetimeindex()
```

```python
S2Sensemble = S2Sensemble.chunk({"member": -1})
```

Grab list of file names.

```python
# files =  !ls /glade/derecho/scratch/jaye/{var}/
files =  !ls /glade/derecho/scratch/mdfowler/temp/{var}/*.nc

# Extracts initialization date from file names.
initializations = [f.split("_")[4] for f in files]
# initializations = [f.split("_")[3] for f in files]
#initializations = [f.split("_")[2] for f in files]
initializations = np.unique(initializations)
```

```python
initializations
```

```python
def update_leads(x):
    """Converts from time coordinates to lead coordinates.
    I.e., lead days.
    """
    x = x.rename({"time": "lead"})
    x = x.assign_coords(lead=np.arange(x.lead.size).astype(int))
    
    return x

def drop_duplicates_along_all_dims(obj, keep=False):
    deduplicated = obj
    for dim in obj.dims:
        indexes = {dim: ~deduplicated.get_index(dim).duplicated(keep=keep)}
        deduplicated = deduplicated.isel(indexes)
    return deduplicated
```

The below loop goes through each initialization and loads in all 10 ensemble members. It converts dimension names to the appropriate names for `climpred` and then chunks them into 100MB chunks in `dask`. We append it all to one big python list which will be used to concatenate into a large dataset with `(init, lead, member, lat, lon)`. This took about 2 minutes for me to run.

```python
S2Sensemble = []
# Loop through each initialization and concatenate.
for init in tqdm(initializations):
    # Sort to make sure we have the same ensemble members at every init.
    filelist = np.sort(
        # glob.glob(filepath + f"{var}_cesm2cam6climoATMv2_{init}_00z_d01_d46*.nc")
        glob.glob(filepath +'/'+ f"{var}_cesm2cam6climoOCNclimoATMv2_{init}_00z_d01_d46*.nc")
    #)[0:5]
    )
    #print(filelist[0])

    # Open all ensemble members for a given init.
    ds = xr.open_mfdataset(
        filelist,
        combine="nested",
        parallel=True,
        concat_dim="member",
        preprocess=drop_duplicates_along_all_dims,
        chunks = {"time": -1, "lon": -1, "lat": -1},
        # speeds things up a bit
        coords="minimal",
        compat="override",
    ).rename({"time": "lead"})

    # Derive datetime for initialization based on the string in the filename.
    MON_TO_INT = {
        "jan": 1,
        "feb": 2,
        "mar": 3,
        "apr": 4,
        "may": 5,
        "jun": 6,
        "jul": 7,
        "aug": 8,
        "sep": 9,
        "oct": 10,
        "nov": 11,
        "dec": 12,
    }

    day = init[0:2]
    mon = init[2:5]
    mon = MON_TO_INT[mon]
    year = init[5::]

    # Assign initialization year and member numbers as coordinates.
    ds = ds.assign_coords(
        lead=np.arange(ds.lead.size).astype(int),
        init=cftime.DatetimeNoLeap(int(year), int(mon), int(day)),
        member=np.arange(ds[var].shape[0]) + 1,
#        member=np.arange(5) + 1,
    )

    # Chunk into one chunk. About 100MB per initialization
    # (full globe, all leads, all members)
    ds = ds.chunk({"member": -1})
    S2Sensemble.append(ds)
    
```

```python
filelist
```

Now we can concatenate into one big dataset. Keep in mind we've used `dask` this whole time so everything is out of memory. (Should take seconds or less)

```python
# Concatenate into one dataset with all inits, leads, members.
# Keywords just speed things up a bit.
ds = xr.concat(
    S2Sensemble,
    dim="init",
    coords="minimal",
    compat="override",
    combine_attrs="override",
)
```

This results in a ~100GB ensemble with 887 inits, 10 members, 45 leads, 180x360 lat/lon. We did all the chunking beforehand so the chunks look good. I'll now save this out as a `zarr` file since it will make bias-correcting much more efficient than trying to work with a netCDF. `zarr` is aware of the chunks we set up here.

```python
ds = ds.sortby("init")
ds['init'] = ds.indexes['init'].to_datetimeindex()
```

```python
if var=="tas_2m":
    ds = ds.rename({"tas_2m": "t2m"})  #For the CESM2 run for consistency
    display(ds.t2m)
elif var=="zg_500":
    ds = ds.rename({"zg_500": "gh_500"})
    display(ds.gh_500)
elif var=="pr_sfc":
    ds = ds.rename ({"pr_sfc": "tp"}); ds['tp'] = ds['tp']*86400
    display(ds.tp)
```
