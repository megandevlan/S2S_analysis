```python
cluster = PBSCluster(
    project='P93300041',
    cores = 1,
    memory = '100GiB',
    processes = 1,
    local_directory = '/glade/derecho/scratch/mdfowler/dask/pbs.$PBS_JOBID/dask/spill',
    resource_spec = 'select=1:ncpus=1:mem=100GB',
    queue = 'casper',
    walltime = '01:00:00',
    interface = 'ext'
)
```
